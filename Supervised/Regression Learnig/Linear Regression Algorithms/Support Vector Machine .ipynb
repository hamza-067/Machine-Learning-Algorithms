```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("placed.csv")
dataset.head(3)
```

```python
dataset= dataset.drop(columns="placed")
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgps",y="score",data=dataset)
plt.show()
```

```python
x = dataset[["cgps"]]
y = dataset["score"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'
```

```python
from sklearn.svm import SVR
```

```python
sr = SVR(kernel="linear")
sr.fit(x_train,y_train)
```

```python
sr.score(x_test,y_test)*100
```

```python
sr.score(x_train,y_train)*100
```

```python
sns.scatterplot(x="cgps",y="score",data=dataset)
plt.plot(dataset["cgps"],sr.predict(x),color="red") 
plt.show()
```
