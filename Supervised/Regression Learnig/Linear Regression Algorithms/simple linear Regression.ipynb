simple linear Regression is type of regression algorithms that model the relationship between a dependent variable 
and single independent variable

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("placement (1).csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["cgpa"]]
x.ndim
```

```python
y = dataset["package"]
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42) 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
6.89	3.26
```

```python
lr.predict([[6.89]])
```

```python
# we will check how many percent give us accuracy 
lr.score(x_test,y_test)*100
```

```python
# 77.30984312051673 this is poor accuracy then if u want increase accuracy then u must be change or increase random_state = numberabs
```
