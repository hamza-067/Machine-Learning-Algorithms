Naive Bayes is a family of simple, probabilistic, supervised machine learning algorithms used for classification tasks. It's based on Bayes' Theorem and makes the "naive" assumption that all features contributing to the outcome are conditionally independent of one another. This assumption simplifies the calculations, making Naive Bayes efficient, easy to build, and effective for very large datasets in applications like spam filtering and sentiment analysis

# Gaussian Naive Bayes:

Use Case:  Best for continuous features that are normally distributed, meaning they form a bell-shaped curve.

How it Works:  It models the probability of features by assuming they follow a Gaussian distribution within each class and estimates the mean and standard deviation of each feature for each clas

# Multinomial Naive Bayes:

Use Case: Suitable for discrete data, particularly when features represent counts or frequencies, like word counts in text classification or spam filtering.

How it Works: Assumes features are generated from a multinomial distribution, making it ideal for situations where a feature can occur multiple time

# Bernoulli Naive Bayes:

Use Case: Used when features are binary (True/False or 1/0), such as the presence or absence of a word in a document (e.g., in a "bag of words" model).

How it Works: Models whether a feature is present or absent in a sample

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset = pd.read_csv("placed.csv")
dataset.head(3)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
#sns.kdeplot(data=dataset['score'])
#plt.show
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgps",y="score",data=dataset,hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2 ,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100 , gnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100, mnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
br = BernoulliNB()
br.fit(x_train,y_train)
```

```python
br.score(x_test,y_test)*100, br.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=br)
plt.show()
```
