support vector machine is one of the most popular supervised learning algorithms, which is used for classification as well as regression problem

# Hard Margin

The algorithms aims to find a hyperplane that perfectly separate the data into two classes without any misclassification.

# Soft Margin

The algorithms allows for some misclassification to find a hyperplane that generalizes to better unseen data and is more rebust to outlier.

# Simple SVM

Typically used for the linear regression and classification problem.

# Kernel SVM

Has more flexibility for non-linear data beacuse u can add more feature to fit a hyperplane instead of a two dimensional space.

# kernel function

# Linear kernel: 
The simplest kernel, equivalent to the original data without transformation. It is used for linearly separable data and creates a flat decision boundary.

# Polynomial kernel:
Maps data into a higher-dimensional space using polynomial equations.

# Radial Basis Function (RBF) kernel:
A powerful and widely used kernel that can handle complex, non-linear data by mapping it to an infinite-dimensional space.

# Sigmoid kernel: 
Uses the sigmoid function (tanh) to map data, often used in neural networks

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("placed.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgps",y="score",data=dataset,hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sc = SVC(kernel="linear")
sc.fit(x_train,y_train)
```

```python
sc.score(x_test,y_test)*100
```

```python
sc.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sc)
plt.show()
```
