# Classification Algorithm

1 . The Classification algorithm is used to identify the category of new observations on the base of training data.

2 . in classification, on program learn from the given dataset or observation and then classifies new observation into a number of classes or group.

3 . such as, Yes or No , 0 or 1, spam or notspam, cat or dog, etc. classes can be called as targets/labels or categories.

# Type of Classification Algroithm

# Binary Classifier:

if the classification problem has only two possible outcome. then it is called binary classifier

Example :  spam or notspam, cat or dog, etc

# Multi-Class Classifier:

if a classification problem has more than outcome, then it is called Multi-Class Classifier

Example : classifications of types of crops, Classification of types musices.
