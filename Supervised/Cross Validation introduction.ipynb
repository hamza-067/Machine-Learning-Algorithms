# Cross Validation

Cross-validation is a resampling procedure in machine learning used to evaluate the performance of a model on a limited data sample and assess how the results of a statistical analysis will generalize to an independent dataset. It is a more robust method for model evaluation compared to a simple train-test split, especially when dealing with smaller dataset

# Method Used for Cross-Validation

1. Leave P out cross-validation

2. Leave one out cross-validation

3. Holdout Cross-validation

4. Repeated Random subsampling validation

5. k-fold cross-validation

6. Stratified K-fold Cross-validation

7. Time Series Cross-validation

8. Nested Cross-Validation

# K-Fold Cross-Validation

K-fold cross-validation is a resampling technique for evaluating machine learning models by splitting a dataset into a number (\(k\)) of equal-sized subsets called folds.

# Stratified K-fold Cross-validation

Stratified k-fold cross-validation is a technique that splits data into 'k' folds, ensuring each fold has approximately the same proportion of classes as the original dataset

This method is a variation of k-fold cross-validation used to address issues with imbalanced datasets, providing a more reliable evaluation of model performance than random splitting

# Leave one out cross-validation

Leave one out cross-validation is an exhaustive cross-validation technique.it is a category of Lp OCV with the case of p=1

Leave-one-out cross-validation (LOOCV) is a machine learning technique that uses each data point as a test set one at a time, while the rest of the data is used for training. This is repeated for every data point, resulting in a final model evaluation that is the average of all the individual test results. It is a form of k-fold cross-validation where the number of folds (\(k\)) is equal to the number of data points

# Leave P out cross-validation

Leave-P-Out cross-validation (LPOCV) is an exhaustive method that trains a model on all possible training sets of size \(n-p\) and tests it on all possible validation sets of size \(p\), where \(n\) is the total number of data points and \(p\) is a user-defined number. This process is repeated for every combination, and the final accuracy is the average of all the individual test scores
