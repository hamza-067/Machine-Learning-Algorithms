# K-Mean Clustering

K-means clustering is an unsupervised machine learning algorithm that groups data points into a pre-defined number (\(k\)) of clusters based on their similarity. It works by iteratively assigning each data point to the nearest cluster centroid (mean) and then recalculating the centroids until the cluster assignments no longer change or a convergence criterion is met.

Method of K-mean clustering 

1. central random

2. Grouping

3. Mean(move)

4. made Perfect cluster Algorithms

5. find the best group to the help of Elbow Method(with cluster sum of squares)

6. Distance Calculate

7. elbow point show best number of cluster (optimal number of clusters)

8. if this method not make best cluster then we use k-mean++


```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset = pd.read_csv("iris.csv")
dataset.head(3)
```

```python
dataset.drop(columns="variety",inplace=True)
```

```python
dataset.head(3)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2,21):
    km = KMeans(n_clusters=i,init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xticks([i for i in range(2,21)])
plt.xlabel("no of cluster")
plt.ylabel("wcss")
plt.grid(axis="x")
plt.show()
```

```python
kmn = KMeans(n_clusters=3)
dataset["Predict"]=kmn.fit_predict(dataset)
```

```python
dataset
```

```python
sns.pairplot(data=dataset,hue="Predict")
plt.savefig("Predict.jpg")
plt.show()
```
