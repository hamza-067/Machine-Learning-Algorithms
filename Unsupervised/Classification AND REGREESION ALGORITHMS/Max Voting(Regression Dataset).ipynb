```python
import pandas as pd 
```

```python
dataset = pd.read_csv("placed.csv")
dataset.head(2)
```

```python
dataset.drop(columns=["placed"],inplace=True)
```

```python
dataset.head(2)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["score"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100, lr.score(x_test,y_test)*100
```

```python
sr = SVR()
sr.fit(x_train,y_train)
sr.score(x_train,y_train)*100, sr.score(x_test,y_test)*100
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100, dt.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("lr1",LinearRegression()),("sr1",SVR()),("dt1",DecisionTreeRegressor())]
```

```python
vr = VotingRegressor(li)
vr.fit(x_train,y_train)
```

```python
vr.score(x_train,y_train)*100, vr.score(x_test,y_test)*100
```

```python
df = {"lr":lr.predict(x_test),"dt":dt.predict(x_test),"sr":sr.predict(x_test),"vr":vr.predict(x_test)}
```

```python
dataset = pd.DataFrame(df)
```
