DBSCAN is a density-based clustering algorithm that groups together points that are closely packed, while marking as noise those points that lie alone in low-density regions

It identifies clusters of arbitrary shapes and is robust to outliers, and it does not require a pre-specified number of clusters. DBSCAN uses two parameters: epsilon (\(\varepsilon \)), which defines the maximum distance between points to be considered neighbors, and MinPts, the minimum number of neighbors a point must have to be considered a "core point

1. mini point = 4

2. epsilon (r) = 0.5

3. core point = no point>-minipoint

4. Binary point = no point < minipoint

5. Noise point (outlier)

```python
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=250,noise=0.05)
```

```python
import pandas as pd
```

```python
df = {"data1":x[:,0] , "data2":x[:,1]}
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.5,min_samples=5)
dataset["Predict"]= db.fit_predict(dataset)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset,hue="Predict")
plt.show()
```
