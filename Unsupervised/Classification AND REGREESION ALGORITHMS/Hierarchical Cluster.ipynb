# Hierarchical Cluster

Hierarchical clustering in machine learning is an unsupervised algorithm that builds a hierarchy of clusters, represented by a tree-like structure called a dendrogram. Unlike algorithms like K-means, it does not require the number of clusters to be specified beforehand, making it ideal for exploring data relationships at different levels of granularity. The two main approaches are agglomerative (bottom-up) and divisive (top-down

# Dendrogram In Hierarchical Cluster

A dendrogram is a tree-like diagram that visualizes the output of hierarchical clustering in machine learning. It displays the merging of individual data points into clusters at different levels of similarity or distance, with the height of the branches representing the distance at which clusters were joined. The dendrogram's structure helps in identifying natural groupings and determining the optimal number of clusters by "cutting" the tree at a chosen height

# Hierarchical clustering technique has two approaches:

# agglomerative

 Agglomerative clustering is a "bottom-up" approach that starts with each data point as its own cluster and merges them based on similarity

#  Divisive

Divisive clustering is a "top-down" approach that begins with all data points in a single cluster and recursively splits it into smaller clusters until each point is in its own cluster

```python
import pandas as  pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("iris.csv")
dataset.head(3)
```

```python
dataset.drop(columns="variety",inplace=True)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,15))
sc.dendrogram(sc.linkage(dataset,method='single', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage="single")
dataset["Predict"] = ac.fit_predict(dataset)
```

```python
dataset.head(3)
```

```python
sns.pairplot(data=dataset,hue="Predict")
plt.show()
```
